```python
import csv
import string
from collections import defaultdict

# Step 1: Define a function to parse and tokenize the text file, remove punctuation, and count words
def process_text(file_path):
    word_counts = defaultdict(int)
    
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read().lower()  # Convert text to lowercase
    
    # Tokenize the text and count word frequencies
    words = text.split()
    translator = str.maketrans('', '', string.punctuation)  # Remove punctuation
    
    for word in words:
        word = word.translate(translator)
        word_counts[word] += 1
    
    return word_counts

# Step 2: Specify the path to your text file
file_path = 'pg5827.txt'  # Replace with the actual file path

# Step 3: Call the function to process the text
word_counts = process_text(file_path)

# Step 4: Create and save the CSV file
csv_file = 'all_words.csv'

with open(csv_file, 'w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(['Word', 'Frequency'])
    
    # Sort the words alphabetically
    sorted_words = sorted(word_counts.keys())
    
    for word in sorted_words:
        csv_writer.writerow([word, word_counts[word]])
```

```python
import csv
import string
from collections import defaultdict

# Step 1: Define a function to tokenize the text file, extract capitalized words, and count their frequencies
def process_text_for_uppercase_words(file_path):
    word_counts = defaultdict(int)
    
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    
    # Tokenize the text and count capitalized word frequencies
    words = text.split()
    translator = str.maketrans('', '', string.punctuation)  # Remove punctuation
    
    for word in words:
        # Check if the word starts with an uppercase letter and contains no lowercase letters
        if word[0].isupper() and not any(char.islower() for char in word):
            word = word.translate(translator)  # Remove punctuation from the word
            word_counts[word] += 1
    
    return word_counts

# Step 2: Specify the path to your text file
file_path = 'pg5827.txt'  # Replace with the actual file path

# Step 3: Call the function to process the text for capitalized words
uppercase_word_counts = process_text_for_uppercase_words(file_path)

# Step 4: Create and save the CSV file for capitalized words
csv_file = 'all_uppercase_words.csv'

with open(csv_file, 'w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(['Word', 'Frequency'])
    
    # Sort the capitalized words alphabetically
    sorted_uppercase_words = sorted(uppercase_word_counts.keys())
    
    for word in sorted_uppercase_words:
        csv_writer.writerow([word, uppercase_word_counts[word]])
```
